#1. 4-connectivity, start position is [0,0], goal position is [5,11]

```python
def neighbor_search(row, col):
  up = [row-1, col]
  down = [row+1, col]
  left = [row, col-1]
  right = [row, col+1]
  if row == 0 and col == 0:
    return down, right
  elif row == 5 and col == 0:
    return up, right
  elif row == 0 and col == 11:
    return down, left
  elif row == 5 and col == 11:
    return up, left
  elif row == 0 and col > 0:
    return down, left, right
  elif row > 0 and col == 0:
    return up, down, right
  elif row == 5 and col > 0:
    return up, left, right
  elif row > 0 and col == 11:
    return up, down, left
  else:
    return up, down, left, right
```

```python
def search_index(value):
  index_list_of_value = []
  for i in range(6):
    for j in range(12):
      if my_matrix_copy.iloc[i,j] == value:
        index_list_of_value.append([i,j])
      else:
        next
  return index_list_of_value
```

```python
def Get_Path(Goal_Position):
  k = (x for x in range(1,100000))
  def get_path(position):
    if search_index(value=0) != []:
      for j in position:
        j_row, j_col = j
        assign_value_equal_zero(row_new=j_row, col_new=j_col)
      p = search_index(value=next(k))
      return get_path(position=p)
    else:
      return my_matrix_copy
  get_path(position=[Goal_Position])
  return my_matrix_copy
```

#2. visualize the path, count the length, final state of the wave expansion

```python
def count_the_length(Start_Position):
  start_row, start_col = Start_Position
  my_dict = {}
  n = neighbor_search(row=start_row, col=start_col)
  for i in n:
    i_row, i_col = i
    if type(my_matrix_copy.iloc[i_row, i_col]) != str:
      my_dict[(i_row,i_col)] = my_matrix_copy.iloc[i_row, i_col]
    else:
      next
  my_min=min(my_dict.values())
  return my_min
```

```python
def visualize_path(Start_Position):
  start_row, start_col = Start_Position
  my_dict = {}
  n = neighbor_search(row=start_row, col=start_col)
  for i in n:
    i_row, i_col = i
    if type(my_matrix_copy.iloc[i_row, i_col]) != str:
      my_dict[(i_row,i_col)] = my_matrix_copy.iloc[i_row, i_col]
    else:
      next
  my_min=min(my_dict, key=my_dict.get)
  row_my_min, col_my_min = my_min
  if my_matrix_copy.iloc[row_my_min, col_my_min] == 1:
    my_matrix_copy.iloc[row_my_min, col_my_min] = '<>'
    return my_matrix_copy
  else:
    my_matrix_copy.iloc[row_my_min, col_my_min] = '<>' 
    return visualize_path(Start_Position=my_min)
```

```python
print('the length of shortest path is', count_the_length([0,0]))
visualize_path(Start_Position=[0,0])
# the <> indicates the shortest path
```

#3. 8-connectivity, start position is [0,0], goal position is [5,11]

```python
#modify the neighbor search as 8-connectivity
def neighbor_search(row, col):
  up = [row-1, col]
  down = [row+1, col]
  left = [row, col-1]
  right = [row, col+1]
  up_right = [row-1, col+1]
  down_right = [row+1, col+1]
  up_left = [row-1, col-1]
  down_left = [row+1, col-1]
  if row == 0 and col == 0:
    return [down, right, down_right]
  elif row == 5 and col == 0:
    return [up, right, up_right]
  elif row == 0 and col == 11:
    return [down, left, down_left]
  elif row == 5 and col == 11:
    return [up, left, up_left]
  elif row == 0 and col > 0:
    return [down, left, right, down_left, down_right]
  elif row > 0 and col == 0:
    return [up, down, right, up_right, down_right]
  elif row == 5 and col > 0:
    return [up, left, right, up_left, up_right]
  elif row > 0 and col == 11:
    return [up, down, left, up_left, down_left]
  else:
    return [up, down, left, right, up_right, down_right, up_left, down_left]
```

```python
def generate_position():
  row = [x for x in range(6)]
  col = [x for x in range(12)]
  Start_Position = [random.sample(row,1)[0], random.sample(col,1)[0]]
  Goal_Position = [random.sample(row,1)[0], random.sample(col,1)[0]]
  obstacle_area = search_index(value='x')
  if Start_Position == Goal_Position:
    return generate_position()
  elif Start_Position in obstacle_area or Goal_Position in obstacle_area:
    return generate_position()
  else:
    return Start_Position, Goal_Position
```

#5. reflection

The drawback of the wavefront algorithm is that it will try each paths to find the shortest path, although it can always generate the global shortest path, 
but it takes a long time to calculate when the map is big. But i find on advantage of the algorithm is that we can set multiple obstacle area in the algorithm, 
the thing that only need to be changed is the margin condition, which is the neighbor_search function here.\
\
My problem here is i did not set a good stoping condition in the Get_Path function, the function has to fill the map with all non-zero value, but it is not efficient 
in the real life. the efficient way is to stop when the 'wave' touchs the Start_Position
